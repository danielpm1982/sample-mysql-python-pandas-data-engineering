## This project exemplifies a simple ETL process, put into practice: 1 - heterogeneous Data is extracted from different local and remote resources; 2 - the Data is treated and structured into the right DB schema (schema.sql); 3 - for loading the final Data into the created local Database, populating .sql files are generated through Python (Pandas) and executed manually at a DBMS GUI client. At the final part of this project, it is shown how to do this loading exclusively through coding. A Python/MySQL driver and associated libs are used then, in order to open a real-time Connection between this Python program and the DBMS.

```python
import pandas as pd
```

```python
book_shop_df = pd.read_csv("books.csv")
book_shop_df.head()

# empty values, at the .csv columns, will be read as 'NaN', by Pandas read_csv() funcion. This "NaN" type will match all empty 
# values when merging two dataframes - as done in the sequence. Either you change the NaN values to an equivalent string (such as 
# "Unknown") before creating the second df from this first one, so that the two dfs would keep matching entries accordingly, or 
# you would let to change - the NaN to a string - only after the merge (join) is done, in order to not hinder the matching of 
# these values between the two dfs. Here, we'll choose this second approach: we'll rename the NaN values only after the merge, 
# but before exporting these values to the populate sql files.
# print(f"{len(book_shop_df[book_shop_df['Genre'].isna()])} rows with empty Genre values, read as NaN:")
# book_shop_df[book_shop_df["Genre"].isna()]
```

```python
unique_author_df = pd.DataFrame(book_shop_df["AUTHOR"].unique(), columns=["AUTHOR"])
unique_author_df["AUTHOR_ID"] = unique_author_df.index + 1
unique_author_df.head(10)
```

```python
book_shop_df = book_shop_df.merge(unique_author_df, on="AUTHOR", how="left")
book_shop_df.head(20)
```

```python
unique_genre_df = pd.DataFrame(book_shop_df["GENRE"].unique(), columns=["GENRE"])
unique_genre_df["GENRE_ID"] = unique_genre_df.index + 1
unique_genre_df
```

```python
book_shop_df = book_shop_df.merge(unique_genre_df, on="GENRE", how="left")
book_shop_df.head(10)
```

```python
unique_author_df = unique_author_df.drop("AUTHOR_ID", axis=1)
unique_author_df

# as with the GENRE_ID from the unique_genre_df, we won't need the AUTHOR_ID from the unique_author_df anymore... therefore, we 
# can simply remove this whole column from the unique_author_df. These ID columns, at the unique_author_df and unique_genre_df 
# were meant to simulate what the DBMS will do by autoincrement, regarding the respective PK values... and to leave a 
# correspondent column at the book_shop_df, as FK (through the merge), in order to stablish the associations at the DB.
```

```python
with open("author_table_populate.sql", "w", encoding="utf-8") as f:
    for _, row in unique_author_df.iterrows():
        name = row["AUTHOR"].replace("'", "''")
        f.write(f"INSERT INTO AUTHOR(NAME) VALUES ('{name}');\n")
```

```python
with open("genre_table_populate.sql", "w", encoding="utf-8") as f:
    for _, row in unique_genre_df.iterrows():
        name = str(row["GENRE"]).replace("'", "''")
        f.write(f"INSERT INTO GENRE(NAME) VALUES ('{name}');\n")
```

```python
with open("book_table_populate.sql", "w", encoding="utf-8") as f:
    for _, row in book_shop_df.iterrows():
        book = row["BOOK"].replace("'", "''")
        language = row["LANGUAGE"].replace("'", "''")
        first_published = int(row["FIRST_PUBLISHED"])
        sales = float(row["SALES (MILLIONS OF US$)"])
        author_id = int(row["AUTHOR_ID"])
        genre_id = int(row["GENRE_ID"])
        sql = f"INSERT INTO BOOK(NAME, LANGUAGE, FIRST_PUBLISHED, SALES, AUTHOR_ID, GENRE_ID) VALUES ('{book}', '{language}',{first_published},{sales:.2f},{author_id},{genre_id});\n"
        f.write(sql)
```

```python
feedback_data = 'https://raw.githubusercontent.com/guilhermeonrails/datas-csv/refs/heads/main/comentarios.json'
feedback_df = pd.read_json(feedback_data)
feedback_df.head(10)
```

```python
feedback_df.shape
```

```python
book_shop_df.head()
```

```python
if "LIVRO_ID" not in feedback_df.columns:
    feedback_df = feedback_df.merge(
        book_shop_df[["BOOK"]].reset_index(drop=False).rename(columns={"index": "LIVRO_ID"}),
        left_on="livro",
        right_on="BOOK",
        how="left"
    )
    feedback_df["LIVRO_ID"] +=1
else:
    print("* This merge has already been done before and the feedback_df already contains LIVRO_ID column. New merge ignored !\n")
feedback_df.head()
```

```python
# Now that the association between the feedback_df and book_shop_df has been defined - through the merge - we no longer need 
# either "livro", "BOOK" or the "id" columns at the feedback_df, only the LIVRO_ID column (other than nome, sobrenome and 
# comentario). LIVRO_ID will be the FK needed for mapping the feedbacks to the correct DB instances at the BOOK table. All PKs,
# including the PK for the FEEDBACK_PT_BR table, will be auto-created by the SGBD (they're not exported from here). The FKs, 
# however, have been created and set here already, between the associated dfs, and will be exported at the sql statements to the 
# actual DB. The simulation used here regarding the PK auto-generation (that only happens at the SGBD), in order to set the 
# corresponding FKs at the associated dfs, necessarily assumes that the auto-generation will follow the same pattern (sequenced 
# int numbers starting from 1). If the auto-generation at the SGBD follows another pattern, all FKs here will 
# point out to the wrong auto-generated PKs. This risk must be clear and well considered by the data engineer and DBAs who will 
# manage the DB squema. Anyone who manages the schema of the DB must know that all PKs must start at 1 and be an int sequence
# from the first until the last registry there, for all tables.

feedback_df = feedback_df.drop("livro", axis=1)
feedback_df = feedback_df.drop("BOOK", axis=1)
feedback_df = feedback_df.drop("id", axis=1)
```

```python
feedback_df.isnull().sum()
```

```python
def format_value(value):
    if pd.isna(value):
        return "NULL"
    elif isinstance(value, str):
        value = value.replace("'","''")
        return f"'{value}'"
    else:
        return str(value)
```

```python
with open("feedback_pt_br_table_populate.sql", "w", encoding="utf-8") as f:
    for _, row in feedback_df.iterrows():
        values = (
            format_value(row["LIVRO_ID"]),
            format_value(row["nome"]),
            format_value(row["sobrenome"]),
            format_value(row["comentario"])
        )
        sql = f"INSERT INTO FEEDBACK_PT_BR(LIVRO_ID, NOME_CLIENTE, SOBRENOME_CLIENTE, FEEDBACK) VALUES ({', '.join(values)});\n"
        f.write(sql)
```

## How to connect to a DB directly from a Python/Pandas program and load Data directly into the MySQL DB tables, as well as retrieve it, as needed. Using SQLAlchemy lib and PyMySQL driver for that.

```python
import sqlalchemy
```

```python
# This function gets a new Connection thread, for the passed db_url, from a pool of connections at the sqlalchemy Engine.
# When using "with" statements for managing resources, these are automatically closed, either the execution beign successful 
# or throwing any error... which avoids memory leaks and also refrains us from explicitly having to use the close() method, on 
# each resource, when the latter is no longer needed (at a "finally" statement, for instance).
# The Connection threads are deallocated and returned to the Engine pool. If one wishes to end the engine with all pooled 
# Connection instances, he should call engine.dispose(), or set no pooling at all when creating the Engine object- by setting 
# "poolclass=NullPool".
# Through Pandas, the Connection thread is used to retrieve the result from the DB, based on that current SELECT DQL statement, 
# and the returned data is automatically stored into a result Dataframe, and returned, differently from when running the DML 
# other statements below - INSERT, UPDATE, DELETE, CALL - with the execute_stmt() function. This is why the two similar functions
# below have been implemented separately: the former retrieves data from the DB into a Dataframe, the latter simply executes DML 
# statements at the DB, with no data or df returned. In the case of Views, and CALL stored procedures that return DB data, these 
# should be executed with the execute_query() and not with the execute_stmt() functions. CALL stored procedures that do not return 
# any data, however, should be executed with the execute_stmt() function instead, as in the case of INSERT, DELETE and UPDATE.

def execute_query(query = "show tables;", db_url = None):
    try:
        engine = sqlalchemy.create_engine(db_url)
        with engine.connect() as conn:
            print(f"Attempting to execute query: {query}")
            result_df = pd.read_sql(query, conn)
            print("Query successfully executed !")
            return result_df
    except Exception as e:
        print("Query NOT executed ! Check out the database URL and query statement for any errors !")
        print(f"You must pass a valid database URL for establishing a connection with the respective DB.\nYou also must pass a valid query to be executed by the DBMS on the DB through this database URL connection.\nPassed db_url: {db_url}\nPassed (or default) query: {query}")
        print(str(e))
```

```python
# This function does the same as the above one, without returning any data from the DB and without returning a Dataframe to
# its caller. Observe that, here, it must be used sqlalchemy.text() to cast the stmt string into an executable SQL stmt.

def execute_stmt(stmt = None, db_url = None):
    try:
        engine = sqlalchemy.create_engine(db_url)
        with engine.connect() as conn:
            print(f"Attempting to execute stmt: {stmt}")
            conn.execute(sqlalchemy.text(stmt))
            conn.commit()
            print("Statement successfully executed !")
    except Exception as e:
        print("Statement NOT executed ! Check out the database URL and the statement for any errors !")
        print(f"You must pass a valid database URL for establishing a connection with the respective DB.\nYou also must pass a valid statement to be executed by the DBMS on the DB through this database URL connection.\nPassed db_url: {db_url}\nPassed statement: {stmt}")
        print(str(e))
```

### Through this later approach, instead of having to export SQL Insert queries at .sql files to populate the Database with the post-treated Data, and run those populating sql files manually at another program (e.g. MySQL Workbench client), we can simply use the MySQL driver and its Connection to insert that Data directly into the DB, iterating and commiting each row of Data from the the Dataframes directly into the respective DB tables. This would be similar to other programming languages' ORM strategies, such as when using EntityManager or Session objects at JPA/Hibernate with Java. Here, the main object managing the transaction flow is the Dataframe itself. There are other ways of implementing this, as well as other libs and drivers. The most compatible way being the one we use here.
